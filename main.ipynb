```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lendo o arquivo csv
data = pd.read_csv('dados_violencia_mulheres_ses_2023.csv', sep=';')
```

```python
filtered_data = data[data['CS_RACA'] != "Ignorado"]
cor_counts = filtered_data['CS_RACA'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(cor_counts.index, cor_counts.values)
plt.title('Etnia das mulheres que receberam algum tipo de violência')
plt.xlabel('Etnia')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo X para melhor visualização
plt.show()
```

```python
filtered_data = data[data['AUTOR_SEXO'] != "Ignorado"]
autor_counts = filtered_data['AUTOR_SEXO'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(autor_counts.index, autor_counts.values)
plt.title('Autor de violências')
plt.ylabel('Quantidade')
plt.xlabel('Autor')

plt.show()
```

```python

filtered_data1 = data.query('VIOL_FISIC != "Não" and VIOL_FISIC != "Ignorado"')
filtered_data2 = data.query('VIOL_PSICO != "Não" and VIOL_PSICO != "Ignorado"')
filtered_data3 = data.query('VIOL_SEXU != "Não" and VIOL_SEXU != "Ignorado"')

concatenated_data = pd.concat([filtered_data1['NU_IDADE_N'],
                               filtered_data2['NU_IDADE_N'],
                               filtered_data3['NU_IDADE_N']],
                              axis=1)

concatenated_data.columns = ['Violência Física', 'Violência Psicológica', 'Violência Sexual']

# Plotando o boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=concatenated_data, color=(50/255, 151/255, 252/255,1), showfliers=False)
plt.title('Boxplot de Violência por Idade')
plt.ylabel('Idade')
plt.show()
```

```python
sns.boxplot(x=data['AUTOR_SEXO'], y=data['NU_IDADE_N'], showfliers=False)
plt.title('Etnia das mulheres que sofreram violência por autor')
plt.xlabel('Autor')
plt.ylabel('Idade')
plt.show()

data = data.dropna(subset=['AUTOR_SEXO'])
autor = data['AUTOR_SEXO']

for sexo in autor.unique():
    plt.figure(figsize=(10, 5))
    sns.histplot(data[data['AUTOR_SEXO'] == sexo]['NU_IDADE_N'], bins=10, stat='frequency')
    plt.xlabel('Idade')
    plt.ylabel('Frequencia(%)')
    plt.title(f'Histograma de Idades para Sexo dos Autores {sexo}')
    plt.show()
```
